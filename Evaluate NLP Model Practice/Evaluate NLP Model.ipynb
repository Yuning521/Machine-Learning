```python
def map_scores_to_labels(x, threshold):
    if (x >threshold):
        return 1
    else:
        return 0

#threshold = 0.5
toxic_predicted = [map_scores_to_labels(x, threshold) for x in df['toxicity_scores']]
```

```python
df['toxic_predicted'] = df.apply(lambda x: map_scores_to_labels(x['toxicity_scores'], 0.5), axis=1)
#df['toxic_predicted'] = [df['toxicity_scores'],0.5].apply(map_scores_to_labels)
```

```python
from sklearn.metrics import classification_report
toxic_list = []

#transfer toxic from {Yes,No} to {0,1} list to compare with toxic_predicted
for i in range(len(df['toxic'])):
    if df['toxic'][i] == 'no':
        toxic_list.append(0)
    else:
        toxic_list.append(1)

#print(toxic_list)       

print (classification_report(toxic_list,df['toxic_predicted']))
```

- Insight: the accuracy of 0.92 is not bad, which means the threshold of 0.5 is not a bad choice, but it's possible that we can choose a better threshold to increase the accuracy.

## Step 3: Get the best threshold by computing the RoC curve 

```python
from sklearn import metrics
fpr, tpr, thresholds = metrics.roc_curve(toxic_list,df['toxicity_scores'])
i = 0
for f in fpr:
    print (f"Threshold = {thresholds[i]}, TPR = {tpr[i]} FPR = {f}")
    i+=1
```

```python
df['toxic_predicted_new'] = df.apply(lambda x: map_scores_to_labels(x['toxicity_scores'], 0.46), axis=1)
print (classification_report(toxic_list,df['toxic_predicted_new']))
```

- Insight: the larger the true positive rate is and the smaller false positive rate is, the better classifier is. So the best threshold here is 0.46 because comparing to the former threshold which is 0.47, the FPR are the same while 0.47 one has lower TPR, and comparing to the later threshold which is 0.45, the TPR are the same while 0.45 has higher FPR. Therefore 0.46 is a turning point which should be the best threshold here. However when we test the threshold = 0.5, the accuracy is 0.92, which is quite close to the accuracy of 0.93. Therefore 0.5 is a good threshold based on the RoC analysis.
